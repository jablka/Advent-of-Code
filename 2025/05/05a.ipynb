```python
s='''3-5
10-14
16-20
12-18

1
5
8
11
17
32
'''
```

```python
with open('input.txt') as f:
    s = f.read()
```

```python
ranges, fresh = s.split('\n\n')
# print(ranges)
# print()
# print(fresh)
```

```python
# print(ranges)
```

```python
fresh_ = fresh.splitlines()
fresh__ = list(map(int,fresh_))
# fresh__
```

```python
ranges_ = ranges.splitlines()
# ranges_
```

```python
ranges__ = []
for i in ranges_:
    start, stop = i.split('-')
    ranges__.append((int(start), int(stop)))
# ranges__
```

```python
# flag = False
result = 0
for a in fresh__:
    flag = False
    for start, stop in ranges__:
        if a>=start and a<=stop:
            result +=1
            flag = True
            break
    if flag:
        continue
result
```
