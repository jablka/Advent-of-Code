```python
result = 0
for arr in lines2:
    maximum = max(arr[:-1])
    position = arr.index(maximum)
    # print(maximum, 'at position:', position)

    new_array = arr[position+1:]
    maximum2 = max(new_array)
    position2 = new_array.index(maximum2)
    # print(maximum2, 'at position:', position2)

    joltage = int(str(maximum)+str(maximum2))
    # print(joltage)

    result += joltage

    # print()
```
