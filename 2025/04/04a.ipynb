```python
with open('input.txt') as f:
    s = f.read()
```

```python
lines = s.splitlines()
# lines
```

```python
x = len(lines[0])
y = len(lines)
# print(x)
# print(y)
```

```python
# now with a few tries
```

```python
papier = 0
nic=0
for r in range(y):
    for sloupec in range(x):
        reference = lines[r][sloupec]
        if reference != '@':
            nic +=1
            continue
        else:
            # print(reference)        
            papier +=1
# print(papier)
# print(nic)
```

```python
counter = 0
interation = 0
for r in range(y):
    for sloupec in range(x):
        reference = lines[r][sloupec]
        if reference != '@':
            continue

        interation +=1

        raz = r-1, sloupec-1
        dva = r-1, sloupec
        tri = r-1, sloupec+1

        styri = r, sloupec-1
        pat = r, sloupec+1

        sest = r+1, sloupec-1
        sedem = r+1, sloupec
        osem = r+1, sloupec+1

        # print([r,sloupec], [raz,dva, tri, styri, pat, sest, sedem, osem])

        result = 0
        for check in [raz, dva, tri, styri, pat, sest, sedem, osem]:
            if (check[0] in range(y)) and (check[1] in range(x)):
                against = lines[check[0]][check[1]]  
            else:
                against = 'nulll'
            # print(against)            
            if against=='@':
                result +=1
        # print('result', result)
        if result > 3:
            continue
        else:
            counter +=1
            # print('counter,',counter)

# print(interation)
print(counter)
```
