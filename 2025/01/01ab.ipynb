```python
with open('input.txt') as f:
    s= f.read()
# s
```

```python
s=s.splitlines()
# s
```

```python
appendix = []
for n in s:
    if n[0]=='L':
        appendix.append(-1*int(n[1:]))
    else:
        appendix.append(int(n[1:]))
# appendix
```

```python
instructions = appendix
```

```python
result = 0
indexer = 50
for step_ in instructions:

    if step_ > 0:
        for i in range(1,abs(step_)+1):
            if indexer+1 > 99:
                indexer = 0
            else:
                indexer+=1
        # print(indexer)

    if step_ < 0:
        for i in range(1,abs(step_)+1):
            if indexer-1 < 0:
                indexer = 99
            else:
                indexer-=1
        # print(indexer)

    if indexer == 0: # check
        result+=1

# print(indexer)
# print()
print(result)
```
