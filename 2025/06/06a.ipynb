```python
s='''123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
'''
```

```python
with open('input.txt') as f:
    s = f.read()
```

```python
lines = s.splitlines()
# lines
```

```python
pocet_riadkov = len(lines)
# pocet_riadkov
```

```python
# s = '123 328  51 64 '
# s.strip()
```

```python
# s = '123 328  51 64 '
# s.split(' ')
```

```python
luggage = []
for i in lines:
    luggage.append(i.split(' '))
# luggage
```

```python
baggage= []
for i in range(len(luggage)-1): # -1 lebo posledný riadok nechceme
    bag = []
    for ii in range(len(luggage[i])):
        if luggage[i][ii]!='':
            bag.append(int(luggage[i][ii]))
    baggage.append(bag)
# baggage
```

```python
znamienka = []
for ii in range(len(luggage[-1])):
    if luggage[-1][ii]!='':
        znamienka.append(luggage[-1][ii])
# znamienka
```

```python
pocet_stlpcov = len(znamienka)
# pocet_stlpcov
```

```python
final_result = 0
for i in range(pocet_stlpcov):

    what = []
    for ii in range(pocet_riadkov-1): # -1 lebo posledný riadok tvorili znamienka
        # print(baggage[ii][i])
        what.append(baggage[ii][i])

    # res = 0
    if znamienka[i] == '*':
        result = 1
        for z in what:
            result *= z

    elif znamienka[i] == '+':
        result = sum(what)

    final_result += result
    # print()
```

```python
print(final_result)
```
